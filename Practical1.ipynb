```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
df=pd.DataFrame({'a':np.random.randint(0,30,100)})
df['b']=df['a']+ np.random.normal(0,20,100)
df['c']=50 - df['a'] + np.random.normal(10,30,100)
df.corr()
pd.plotting.scatter_matrix(df,figsize=(6,6))
plt.show()
```
